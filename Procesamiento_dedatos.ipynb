Analizo las 5 provincias con mas hechos delictivos, y su evolucion en los ultimos 5 años

```python
import numpy as np

# Obtener las cinco provincias con mayor índice de delitos registrados
provincias_top5 = df_reducido['provincia'].value_counts().nlargest(5).index

# Filtrar el DataFrame para incluir solo las provincias top 5
df_top5 = df_reducido[df_reducido['provincia'].isin(provincias_top5)]

# Agrupar los datos por provincia y año y contar los delitos
grupo_provincia_anio = df_top5.groupby(['provincia', 'anio'])['tipo_hecho_segun_victima'].count().reset_index()

# Crear un gráfico de barras agrupados para mostrar la evolución del delito por año para cada provincia
plt.figure(figsize=(12, 8))
plt.title('Evolución del Delito por Año - Provincias Top 5')

# Obtener la cantidad de provincias y años
num_provincias = len(provincias_top5)
num_anios = grupo_provincia_anio['anio'].nunique()

# Calcular el ancho de cada barra y la separación entre grupos
bar_width = 0.8 / num_provincias
group_padding = 0.1

# Crear una lista de índices para las posiciones de las barras
bar_positions = np.arange(num_anios)

# Iterar sobre las provincias top 5 y graficar las barras agrupadas
for i, provincia in enumerate(provincias_top5):
    datos_provincia = grupo_provincia_anio[grupo_provincia_anio['provincia'] == provincia]
    # Calcular la posición horizontal para cada barra en el grupo
    bar_position = bar_positions + i * bar_width
    plt.bar(bar_position, datos_provincia['tipo_hecho_segun_victima'], bar_width, label=provincia)

# Configurar el gráfico
plt.xlabel('Año')
plt.ylabel('Cantidad de Delitos')
plt.xticks(bar_positions + bar_width * (num_provincias - 1) / 2, grupo_provincia_anio['anio'].unique())
plt.legend()
plt.grid(True)
plt.show()
```
